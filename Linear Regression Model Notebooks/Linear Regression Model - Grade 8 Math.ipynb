```python
# Importing libraries.

import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
avg_scores = pd.read_csv('input_data/naep.csv')
poverty_income = pd.read_csv('input_data/poverty_income.csv')
unemploy_crime = pd.read_csv('input_data/unemployment_crime.csv')
```

```python
unemploy_crime.info()
```

```python
merged_df = pd.merge(avg_scores, poverty_income, how = 'left' , left_on =['state', 'year'], right_on=['state', 'year'])
merged_df = pd.merge(merged_df, unemploy_crime, how = 'left' , left_on =['state', 'year'], right_on=['state', 'year'])
merged_df
```

```python
math_merged_df = merged_df.loc[(merged_df['subject'] == 'Mathematics')]
math_8_merged_df = math_merged_df.loc[(merged_df['grade'] == 8)]
math_8_merged_df
```

```python
math_8_merged_df = math_8_merged_df.drop(columns = 'year')
math_8_merged_df = math_8_merged_df.drop(columns = 'state')
math_8_merged_df = math_8_merged_df.drop(columns = 'subject')
math_8_merged_df = math_8_merged_df.drop(columns = 'grade')
math_8_merged_df = math_8_merged_df.drop(columns = 'violent_crime_rate')
math_8_merged_df = math_8_merged_df.drop(columns = 'property_crime_rate')
math_8_merged_df
```

```python
math_8_merged_df.isnull().sum()
```

```python
math_8_merged_df.info()
```

## Linear Regression Testing - Grade 8 Math

```python
# Divide the Data into Attributes & Labels

x = math_8_merged_df.drop(columns = 'avg_score') 
y = math_8_merged_df['avg_score'].values
```

```python
# Split Data into Training & Test Sets.

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 42)
```

```python
# Train the Algorithm.

from sklearn.linear_model import LinearRegression  
regression = LinearRegression()  
regression.fit(x_train, y_train) 
```

```python
# Predict Scores for the Model.

print(x_test) 
y_pred = regression.predict(x_test)
```

```python
# Compare Actual Data vs Predicted Data

score_vs_factors = pd.DataFrame({"Actual Scores": y_test, "Predicted Scores": y_pred})
score_vs_factors
```

```python
pd.DataFrame(np.c_[x_test,y_test,y_pred], columns=["Poverty Percentage","Median Income", "Unemployment Rate","Actual Scores","Predicted Scores"])
```

```python
# Test with Custom Data.

input = [[15.6, 51771, 3.0]]
prediction = regression.predict(input)
print("Predicted Score = {}".format(prediction[0]))
```

```python
# Evaluate Performance of Algorithm.

from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred))
```
